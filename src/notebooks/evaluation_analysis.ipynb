```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
column_names = ['second', 'function_id', 'total_inference_time', 'model_to_gpu_time', 'inference_time', 'unload_from_gpu_time']
baseline = pd.read_csv('../../results/baseline2/evaluation_results.txt', header=None, names=column_names)
column_names = ['second', 'function_id', 'total_inference_time', 'preloaded_function_id_model', 'model_on_gpu_before_preloading', 'manager_time_series_time', 'model_to_gpu_time', 'inference_time', 'unload_from_gpu_time']
experiment_3 = pd.read_csv('../../results/experiment3/evaluation_results.txt', header=None, names=column_names)
column_names = ['second', 'function_id', 'total_inference_time', 'preloaded_function_id_model_1', 'preloaded_function_id_model_2', 'preloaded_function_id_model_3', 'preloaded_function_id_model_4', 'model_on_gpu_before_preloading_1', 'model_on_gpu_before_preloading_2', 'model_on_gpu_before_preloading_3', 'model_on_gpu_before_preloading_4', 'manager_time_series_time', 'model_to_gpu_time', 'inference_time', 'unload_from_gpu_time']
experiment_4 = pd.read_csv('../../results/experiment4/evaluation_results.txt', header=None, names=column_names)
```

```python
baseline['total_inference_time_including_queue'] = 0.0
prev_second = None
cumulative_time = 0.0

for index, row in baseline.iterrows():
    # If the second id changes, reset cumulative time
    if prev_second != row['second']:
        cumulative_time = 0.0
        prev_second = row['second']
    
    # Update cumulative time with the current row's inference time
    cumulative_time += row['total_inference_time']
    
    # Update the new column with the cumulative time
    baseline.at[index, 'total_inference_time_including_queue'] = cumulative_time

experiment_3['total_inference_time_including_queue'] = 0.0
prev_second = None
cumulative_time = 0.0

for index, row in experiment_3.iterrows():
    # If the second id changes, reset cumulative time
    if prev_second != row['second']:
        cumulative_time = 0.0
        prev_second = row['second']
    
    # Update cumulative time with the current row's inference time
    cumulative_time += row['total_inference_time']
    
    # Update the new column with the cumulative time
    experiment_3.at[index, 'total_inference_time_including_queue'] = cumulative_time

experiment_4['total_inference_time_including_queue'] = 0.0
prev_second = None
cumulative_time = 0.0

for index, row in experiment_4.iterrows():
    # If the second id changes, reset cumulative time
    if prev_second != row['second']:
        cumulative_time = 0.0
        prev_second = row['second']
    
    # Update cumulative time with the current row's inference time
    cumulative_time += row['total_inference_time']
    
    # Update the new column with the cumulative time
    experiment_4.at[index, 'total_inference_time_including_queue'] = cumulative_time
```
