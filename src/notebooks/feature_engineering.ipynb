# Count how many times each function was invoked

```python
invocation_count = pd.DataFrame(columns=['app', 'func', 'invocation_count'])

for app in tqdm(data_raw["app"].unique()):
    filtered_df = data_raw.loc[data_raw['app'] == app]
    for func in filtered_df["func"].unique():
        df_to_count = data_raw[(data_raw['app'] == app) & (data_raw['func'] == func)]
        row_count = len(df_to_count)
        invocation_count = pd.concat([invocation_count, pd.DataFrame({'app': [app], 'func': [func], 'invocation_count': [row_count]})], ignore_index=True)
```

```python
invocation_count = invocation_count.sort_values(by='invocation_count', ascending=False)
invocation_count
```

# Iterate over each function, do feature engineering and save it's data in a file

```python
counter = 1
for index, row in tqdm(invocation_count.iterrows()):

    # generate data for only 100 most frequently used functions
    if counter > 100:
        break

    app = row["app"]
    func = row["func"]
    function = data_raw[(data_raw['app'] == app) & (data_raw['func'] == func)]
    function['second'] = function['start_timestamp'].astype(int)

    # create values (invocation=0) for missing datapoints
    #function = function[["second"]]
    unique_values = function['second'].unique()
    counts = function['second'].value_counts()
    new_range = range(0, 1209601)
    new_values = [x for x in new_range if x not in unique_values]
    new_counts = [0] * len(new_values)

    new_data = {'second': unique_values.tolist() + new_values, 'invocations': counts.tolist() + new_counts}
    new_df = pd.DataFrame(new_data)

    new_df.reset_index(drop=True, inplace=True)
    new_df = new_df.sort_values(by='second', ascending=True)
    new_df

    # convert invocation values -> either 0 or 1
    invocations = new_df["invocations"]
    binary_list = [1 if value >= 1 else 0 for value in invocations]
    new_df["invocations"] = binary_list
    new_df

    # FEATURE ENGINEERING ####################################################

    # Create 'time_since_prev_invocation' feature - time passed since last invocation (in seconds)
    seconds_since_last_invocation = []
    last_invocation_time = 1
    # Iterate through the DataFrame
    for index, row in new_df.iterrows():
        if row['invocations'] == 1:
            seconds_since_last_invocation.append(last_invocation_time)
            last_invocation_time = 1
        else:
            seconds_since_last_invocation.append(last_invocation_time)
            last_invocation_time += 1

    new_df['time_since_prev_invocation'] = seconds_since_last_invocation
    new_df

    function.drop(columns=["app", "func", "duration", "start_timestamp"], inplace=True)
    all_data = pd.merge(new_df, function, on='second', how="left")

    start_date = datetime(2021, 1, 31)
    all_data['actual_timestamp'] = all_data['second'].apply(lambda x: start_date + timedelta(seconds=x))

    # Extract date and time features
    all_data['year'] = all_data['actual_timestamp'].dt.year
    all_data['month'] = all_data['actual_timestamp'].dt.month
    all_data['day'] = all_data['actual_timestamp'].dt.day
    all_data['hour'] = all_data['actual_timestamp'].dt.hour
    all_data['minute'] = all_data['actual_timestamp'].dt.minute
    all_data['second_of_day'] = all_data['actual_timestamp'].dt.hour * 3600 + all_data['actual_timestamp'].dt.minute * 60 + all_data['actual_timestamp'].dt.second
    all_data['is_weekday'] = (all_data['actual_timestamp'].dt.weekday < 5).astype(int)  # 1 for weekday, 0 for weekend

    # The below doesn't help and worsenes the results by a lot
    # Encode the month, day, hour, minute and second_of_day as sine and cosine components
    # (cyclic encoding for periodic features)
    #all_data['month_sin'] = np.sin(2 * np.pi * all_data['month'] / 12)
    #all_data['month_cos'] = np.cos(2 * np.pi * all_data['month'] / 12)
    #all_data['hour_sin'] = np.sin(2 * np.pi * all_data['hour'] / 24)
    #all_data['hour_cos'] = np.cos(2 * np.pi * all_data['hour'] / 24)
    #all_data['minute_sin'] = np.sin(2 * np.pi * all_data['minute'] / 60)
    #all_data['minute_cos'] = np.cos(2 * np.pi * all_data['minute'] / 60)

    #all_data.drop(columns=["month", "hour", "minute"], inplace=True)

    # The below doesn't help and worsenes the results by a lot
    # Calculate rolling window statistics: mean, sum, and standard deviation for 'invocations'
    #window_size = 1000
    #df_shifted = all_data.shift(1) # do not consider current timestep data
    #all_data['invocations_mean'] = df_shifted['invocations'].rolling(window=window_size).mean()
    #all_data['invocations_sum'] = df_shifted['invocations'].rolling(window=window_size).sum()
    #all_data['invocations_std'] = df_shifted['invocations'].rolling(window=window_size).std()

    all_data.drop(columns=["actual_timestamp"], inplace=True)
    all_data.to_csv(f"../../data/training_data/{counter}.txt")
    counter += 1
```
